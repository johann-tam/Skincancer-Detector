```python
import numpy as np
import pandas as pd
import os


from PIL import Image
from django.conf import settings

import tensorflow as tf
```

```python
width = 28
height = 28

# variables
columns = width * height

# Array hold the headings
number_array = np.arange(columns)
string_array = []
```

```python
for i in range(number_array.size):
   string_array.append("pixel" + str(np.char.zfill(str(i), 4)))


string_array.append("label")

print(string_array)
```

```python
df = pd.DataFrame(columns=string_array)
df.head()
```

```python
# Get pixelvalues
def process_jpg(filepath):
    # load image and convert it to greyscale
    img = Image.open(filepath).convert('L')
    # resize the image to the specified width and height that the dl_model was trained on
    resized_img = img.resize((width, height))
    # get the pixel values for the image
    image_sequence = resized_img.getdata()
    # store the pixel values in an numpy array for the dl_model to use
    image_array = np.array(image_sequence)

    print(image_array.size)
    return image_array
```

```python
# Needs folders for each category
# Generated by torchvision-hugsvision-on-ham10000-88-acc.ipynb
# Folders are wrong though, need to be relabelled
# Takes around half an hour to finish

# Goes in each folder of images
for idx, current_dir in enumerate(["./HAM10000/akiec", "./HAM10000/bcc", "./HAM10000/bkl", "./HAM10000/df", "./HAM10000/mel", "./HAM10000/nv", "./HAM10000/vasc"]):
    print(idx)
    # Goes trough each file in the folder
    for image_path in os.listdir(current_dir):
        # exclude .DS_STORE files (MACOS)
        if image_path == current_dir + '.DS_Store':
            continue

        # get pixel data for each image
        pixels = process_jpg(current_dir +'/'+ image_path)


        # Add label to pixel data
        pixels = np.append(pixels, idx)

        print("pixels: " + str(pixels.size))
        # Add pixel data to dataframe
        df = df.append(pd.DataFrame([pixels], columns=df.columns), ignore_index=True)

    # Export to csv for each category
    print(df.shape)
    df.to_csv('RGB_28x28-' + str(idx) + '-grey.csv', index=False)
    # Define dataframe with amount of columns as specified in number_array
    df = pd.DataFrame(columns=string_array)
    print(df.shape)
```
